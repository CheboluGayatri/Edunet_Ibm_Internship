**Working mechanism of ml**


**1.Get Data** : First, we collect data. This data is used to teach the machine. The better and larger the data, the better the model can learn.

In this step, we give the prepared data to a machine learning algorithm. The model learns patterns and relationships from the data.

**5.Improve**

If the results are not good, we improve the model by adding more data, changing the algorithm, or adjusting settings. This process continues until the model performs well.

**Machine Learning Types**

**1. Supervised Learning**

Supervised learning works with labeled data. This means the data contains both input and correct output.

The model learns by comparing its predictions with the correct answers.

Example:
Training a model using fruit names and their types like apple, banana, orange.

Flow:
Train data → ML algorithm → Model → Prediction

Supervised learning has two types:

***a. Classification**
Classification works with categories or classes.
The output is a label.

Examples:
Spam or not spam
Fruit is apple or banana
Pass or fail

**b. Regression**
Regression works with continuous values.
The output is a number.

Examples:
Predicting price of a house
Predicting temperature
Predicting salary

**2. Unsupervised Learning**

Unsupervised learning works with unlabeled data.
There is no predefined output. The model finds patterns on its own.

Examples:
Grouping similar customers
Clustering similar products
Finding hidden patterns in data

Common types include clustering and association.

**Dimensionality Reduction**

Dimensionality reduction is a technique used to reduce the number of input features (variables) in a dataset while keeping the important information.

In simple words, it removes unnecessary or less important features so the model becomes faster and easier to understand.

**Why Dimensionality Reduction is Needed**

Reduces data size

Improves model performance

Removes noise and redundant features

Makes data easier to visualize

Reduces overfitting
eg: good student in academic

**3.Reinforcement Learning**

Reinforcement learning is based on reward and punishment.
The model learns by interacting with an environment and improving its actions over time.

Examples:
Game playing bots
Robot movement
Self driving cars


**Classification:**

classification is a supervised machine learning method where the model tries to predict the correct label of a given input data

in classification,the model is fully trained using the train data,and then it is evaluated on test before being used to perform prediction on new unseen data.

**Types**

**1. Binary Classification**

There are only two possible classes.

Examples:
Yes or No
Spam or Not Spam
Pass or Fail

**2. Multiclass Classification**

There are more than two classes, but only one class is chosen as output.

Examples:
Fruit classification (apple, banana, orange)
Digit recognition (0–9)
Grade prediction (A, B, C)

**3. Multi-Label Classification**

One input can belong to multiple classes at the same time.

Examples:
Movie genres (action, comedy, drama)
Email with multiple tags
Music classification

**4. Imbalanced Classification**

Classes are not equally distributed.

Examples:
Fraud detection
Disease detection
Rare event prediction

**5. Hierarchical Classification**

Classes are arranged in a parent–child structure.

Examples:
Animal classification
Document classification by topics
Product categories

**Regression**

Regression is used to predict numerical values.

Examples:
House price, salary, temperature, marks

**Types of Regression**

**1. Linear Regression**

Predicts output using one input with a straight line relationship.

Example:
Predicting house price using house size.

**2. Multiple Linear Regression**

Predicts output using more than one input.

Example:
House price using size, location, and number of rooms.

**3. Polynomial Regression**

Used when data does not follow a straight line.

Example:
Predicting growth of plants over time.

**4. Logistic Regression**

Used for classification, not numbers.

Example:
Pass or fail
Spam or not spam

**Summary**

Linear: one input → number

Multiple Linear: many inputs → number

Polynomial: curved relationship

Logistic: Yes or No output

**Market Basket Analysis**

Market Basket Analysis is a data analysis technique used to understand customer buying behavior by finding relationships between products purchased together.

In simple words, it tells us “If a customer buys this item, what else are they likely to buy?”

Example

If customers often buy bread and butter together, a store can place them near each other or offer combo discounts.

Another example:
If someone buys milk, they may also buy eggs and bread.

**How It Works**

It analyzes past sales data and finds common item combinations using association rules.

Flow:
Transaction data → Analysis → Product relationships → Business decisions

Common Measures

Support: How often items appear together

Confidence: How likely item B is bought when item A is bought

Lift: Strength of the relationship between items

Uses of Market Basket Analysis

Product placement in stores

Cross selling and recommendations

Discount and promotion planning
